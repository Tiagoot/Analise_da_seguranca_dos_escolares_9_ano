#Reconhecimento de Imagens

 * Utilizando-se de câmeras de vigilância que capturam imagens em tempo real e as comparam com bancos de dados.
 * O reconhecimento facial é usado para identificar indivíduos com base em características faciais únicas.
 * O reconhecimento de placas de veículos é outra aplicação comum.
